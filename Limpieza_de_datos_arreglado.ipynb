```python
import re
def edad_a_dias(texto):
    # Extraer números usando regex
    match = re.findall(r'(\d+)', texto)
    if len(match) == 3:
        años, meses, dias = map(int, match)
        # Aproximación: 1 año = 365 días, 1 mes = 30 días
        return años * 365 + meses * 30 + dias
    return None

# Crear nueva columna
df["Edad_dias"] = df["Edad a fecha de registro"].apply(edad_a_dias)
df['Edad_dias']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier

def imputar_columna_categorica(df, columna_objetivo):
    # Dividir en datos conocidos y desconocidos
    df_known = df[df[columna_objetivo].notnull()]
    df_missing = df[df[columna_objetivo].isnull()]

    if df_missing.empty:
        return df  # no hay nada que imputar

    X_known = df_known.drop(columns=[columna_objetivo])
    y_known = df_known[columna_objetivo]

    X_missing = df_missing.drop(columns=[columna_objetivo])

    # OneHotEncoding con la versión nueva de sklearn
    ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    X_known_encoded = ohe.fit_transform(X_known)
    X_missing_encoded = ohe.transform(X_missing)

    # Modelo para imputar
    model = RandomForestClassifier()
    model.fit(X_known_encoded, y_known)

    # Predecir valores faltantes
    y_pred = model.predict(X_missing_encoded)

    # Asignar valores imputados
    df.loc[df[columna_objetivo].isnull(), columna_objetivo] = y_pred

    return df

```

```python
import seaborn as sns
import altair as alt

#Función para todas
def graficar_barras_apiladas_porcentaje(df, columna_x, columna_hue):

    # Tabulación cruzada
    tabla_cruzada = pd.crosstab(df[columna_x], df[columna_hue])

    # Calcula los porcentajes
    total = tabla_cruzada.sum(axis=1)
    porcentajes = tabla_cruzada.div(total, axis=0) * 100

    #Clases
    ax = porcentajes.plot(kind='bar', stacked=True, figsize=(10, 6))

    #Gráfico
    plt.title(f'Porcentaje {columna_hue} por {columna_x}')
    plt.xlabel(columna_x)
    plt.ylabel('Porcentaje')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title=columna_hue, bbox_to_anchor=(1.05, 1), loc='upper left')

    #Porcentajes
    for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy()
        ax.text(x + width/2,
                y + height/2,
                '{:.0f}%'.format(height),
                horizontalalignment='center',
                verticalalignment='center')


    #Gráfica
    plt.show()

graficar_barras_apiladas_porcentaje(df, 'Sexo', 'Alta General')
```
